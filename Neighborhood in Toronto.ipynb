### Scraping the wikipedia page

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
print(soup.text)
```

```python
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
#extract all the hyperlinks within the webpage.
soup.find_all('a')
```

```python
#use a for loop and the get('"href") method to extract and print out only hyperlinks
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# Print the first 10 rows for sanity check
rows = soup.find_all('tr')
print(rows[:10])
```

```python
#convert the list into a dataframe and get a quick view of the first 10 rows using Pandas.
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

### Data Manipulation and Cleaning

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']
```

```python
df1.head(10)
```

```python
# Combining the neighbourhoods with same Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)
df2
```

```python
# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])
df2
```

```python
df2.shape
```

### Importing Geocoder package for latitude and longitude

```python
geo_loc=pd.read_csv('https://cocl.us/Geospatial_data')
geo_loc.head()
```

### Merging two tables

```python
df3 = pd.merge(df2,geo_loc,)
df3.head()
```

### Only boroughs that contain the word Toronto 

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```
